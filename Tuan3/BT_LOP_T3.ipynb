```python
import copy
import math

#  CONSTANT
X = "X"
O = "O"
EMPTY = None

# BOARD
def initial_state():
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]

def print_board(board):
    for row in board:
        print([" " if cell is None else cell for cell in row])
    print()

#  LOGIC
def player(board):
    count = sum(row.count(X) + row.count(O) for row in board)
    return X if count % 2 == 0 else O

def actions(board):
    return [(i, j)
            for i in range(3)
            for j in range(3)
            if board[i][j] is EMPTY]

def result(board, action):
    new_board = copy.deepcopy(board)
    i, j = action
    new_board[i][j] = player(board)
    return new_board

#  CHECK WIN
def winner(board):
    lines = []

    # Rows
    lines.extend(board)

    # Columns
    lines.extend([[board[i][j] for i in range(3)] for j in range(3)])

    # Diagonals
    lines.append([board[i][i] for i in range(3)])
    lines.append([board[i][2 - i] for i in range(3)])

    for line in lines:
        if line == [X, X, X]:
            return X
        if line == [O, O, O]:
            return O
    return None

def terminal(board):
    return winner(board) is not None or all(
        cell is not EMPTY for row in board for cell in row
    )

def utility(board):
    w = winner(board)
    if w == X:
        return 1
    elif w == O:
        return -1
    return 0

#  MINIMAX
def max_value(board):
    if terminal(board):
        return utility(board)

    v = -math.inf
    for action in actions(board):
        v = max(v, min_value(result(board, action)))
    return v

def min_value(board):
    if terminal(board):
        return utility(board)

    v = math.inf
    for action in actions(board):
        v = min(v, max_value(result(board, action)))
    return v

def minimax(board):
    if terminal(board):
        return None

    turn = player(board)

    if turn == X:
        best_val = -math.inf
        best_move = None
        for action in actions(board):
            val = min_value(result(board, action))
            if val > best_val:
                best_val = val
                best_move = action
    else:
        best_val = math.inf
        best_move = None
        for action in actions(board):
            val = max_value(result(board, action))
            if val < best_val:
                best_val = val
                best_move = action

    return best_move

#  MAIN GAME
def play():
    board = initial_state()

    print("Chọn quân (X đi trước / O đi sau): ")
    human = input().upper()

    if human not in [X, O]:
        print("Chọn X hoặc O!")
        return

    ai = O if human == X else X

    print("\nBắt đầu game!\n")

    while True:
        print_board(board)

        if terminal(board):
            w = winner(board)
            if w is None:
                print("Hòa!")
            else:
                print(f"{w} thắng!")
            break

        turn = player(board)

        if turn == human:
            print("Nhập nước đi (row col): ")
            i = int(input("Row (0-2): "))
            j = int(input("Col (0-2): "))

            if (i, j) in actions(board):
                board = result(board, (i, j))
            else:
                print("Nước đi không hợp lệ!")
        else:
            print("AI đang suy nghĩ...")
            move = minimax(board)
            board = result(board, move)

#  RUN

if __name__ == "__main__":
    print("==============BÀI 1==============2")
    play()
```

```python
import os
import math

# BOARD
def print_board(board):
    os.system('cls' if os.name == 'nt' else 'clear')
    print(f"""
    {board[0]} | {board[1]} | {board[2]}
    ---------
    {board[3]} | {board[4]} | {board[5]}
    ---------
    {board[6]} | {board[7]} | {board[8]}
    """)

def get_winner(board):
    win_states = [
        [0,1,2],[3,4,5],[6,7,8], # rows
        [0,3,6],[1,4,7],[2,5,8], # cols
        [0,4,8],[2,4,6]          # diagonals
    ]

    for a,b,c in win_states:
        if board[a] == board[b] == board[c]:
            return board[a]
    return None

def get_empty_cells(board):
    return [i for i in range(9) if board[i] not in ["X","O"]]

# ALPHA-BETA
def minimax(board, depth, alpha, beta, is_max):
    winner = get_winner(board)

    # Terminal
    if winner == "X":
        return 10 - depth
    elif winner == "O":
        return -10 + depth
    elif len(get_empty_cells(board)) == 0:
        return 0

    if is_max:
        best = -math.inf
        for i in get_empty_cells(board):
            temp = board[i]
            board[i] = "X"

            val = minimax(board, depth+1, alpha, beta, False)
            board[i] = temp

            best = max(best, val)
            alpha = max(alpha, val)

            if beta <= alpha:
                break

        return best

    else:
        best = math.inf
        for i in get_empty_cells(board):
            temp = board[i]
            board[i] = "O"

            val = minimax(board, depth+1, alpha, beta, True)
            board[i] = temp

            best = min(best, val)
            beta = min(beta, val)

            if beta <= alpha:
                break

        return best

#  BEST MOVE
def find_best_move(board, ai):
    best_move = -1

    if ai == "X":
        best_val = -math.inf
        for i in get_empty_cells(board):
            temp = board[i]
            board[i] = "X"

            val = minimax(board, 0, -math.inf, math.inf, False)
            board[i] = temp

            if val > best_val:
                best_val = val
                best_move = i

    else:
        best_val = math.inf
        for i in get_empty_cells(board):
            temp = board[i]
            board[i] = "O"

            val = minimax(board, 0, -math.inf, math.inf, True)
            board[i] = temp

            if val < best_val:
                best_val = val
                best_move = i

    return best_move

# GAME
def main():
    player = input("Chọn X hoặc O: ").upper()
    ai = "O" if player == "X" else "X"

    board = list(range(1,10))
    turn = "X"

    while True:
        print_board(board)

        # AI
        if turn == ai:
            print("AI đang chơi...")
            move = find_best_move(board, ai)
            board[move] = ai
            turn = player

        # Player
        else:
            move = int(input("Nhập vị trí (1-9): ")) - 1
            if board[move] not in ["X","O"]:
                board[move] = player
                turn = ai
            else:
                print("Ô đã chọn!")
                continue

        # Check win
        winner = get_winner(board)
        if winner:
            print_board(board)
            print(f"{winner} thắng!")
            break

        # Check tie
        if len(get_empty_cells(board)) == 0:
            print_board(board)
            print("Hòa!")
            break

if __name__ == "__main__":
    print("==============BÀI 2==============")
    main()
```
