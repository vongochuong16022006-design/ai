```python
#-------cau1.--------
def tron_mang_min(a, b):
    c = []
    n = min(len(a), len(b))

    # Lấy min của các phần tử cùng vị trí
    for i in range(n):
        c.append(min(a[i], b[i]))

    # Thêm phần tử còn lại của mảng dài hơn
    c.extend(a[n:])
    c.extend(b[n:])

    return c

# Chương trình chính
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

print("=======Cau1========")
ket_qua = tron_mang_min(a, b)
print("Mảng kết quả:", ket_qua)
```

```python
#-----------cau2-------------------------
print("=======Cau2========")
# a. Tổng tam giác trên đường chéo phụ
def tong_tam_giac_tren_cheo_phu(a):
    n = len(a)
    tong = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:
                tong += a[i][j]
    return tong


# b. Đổi số âm thành trị tuyệt đối
def doi_so_am(a):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] < 0:
                a[i][j] = abs(a[i][j])


# c. Thay số chẵn bằng x
def thay_chan_bang_x(a, x):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] % 2 == 0:
                a[i][j] = x


# d. Kiểm tra toàn chẵn
def toan_chan(a):
    for hang in a:
        for x in hang:
            if x % 2 != 0:
                return False
    return True


# e. Kiểm tra ma trận đối xứng
def doi_xung(a):
    n = len(a)
    for i in range(n):
        for j in range(n):
            if a[i][j] != a[j][i]:
                return False
    return True


# f. Kiểm tra đường chéo chính tăng dần
def cheo_chinh_tang(a):
    n = len(a)
    for i in range(n - 1):
        if a[i][i] > a[i + 1][i + 1]:
            return False
    return True


# g. Xuất tam giác dưới đường chéo phụ
def xuat_tam_giac_duoi_cheo_phu(a):
    n = len(a)
    for i in range(n):
        for j in range(n):
            if i + j >= n - 1:
                print(a[i][j], end=" ")
        print()


# h. Kiểm tra đường chéo phụ giảm dần
def cheo_phu_giam(a):
    n = len(a)
    for i in range(n - 1):
        if a[i][n - 1 - i] < a[i + 1][n - 2 - i]:
            return False
    return True


# ================= TEST =================
a = [
    [1, -2, 3],
    [4, 5, -6],
    [7, 8, 9]
]

print("\na) Tổng tam giác trên chéo phụ:", tong_tam_giac_tren_cheo_phu(a))

doi_so_am(a)
print("\nb) Sau khi đổi số âm:", a)

thay_chan_bang_x(a, 100)
print("\nc) Sau khi thay số chẵn:", a)

print("\nd) Toàn chẵn:", toan_chan(a))
print("\ne) Đối xứng:", doi_xung(a))
print("\nf) Chéo chính tăng:", cheo_chinh_tang(a))

print("\ng) Tam giác dưới chéo phụ:")
xuat_tam_giac_duoi_cheo_phu(a)

print("\nh) Chéo phụ giảm:", cheo_phu_giam(a))

```
