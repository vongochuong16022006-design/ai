```python
#---------------cau13--------
print("=======cau13========")
def tron_mang_tong(a, b):
    c = []
    n = min(len(a), len(b))

    # Cộng các phần tử cùng vị trí
    for i in range(n):
        c.append(a[i] + b[i])

    # Thêm phần còn lại của mảng dài hơn
    c.extend(a[n:])
    c.extend(b[n:])

    return c
    # Chương trình chính
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

ket_qua = tron_mang_tong(a, b)
print("Mảng kết quả:", ket_qua)
```

```python
#==========Câu 14=================
import random

# a. Tạo ma trận ngẫu nhiên
def tao_ma_tran(m, n):
    a = []
    for i in range(m):
        row = []
        for j in range(n):
            row.append(random.randint(1, 9))
        a.append(row)
    return a

# In ma trận
def xuat_ma_tran(a):
    for row in a:
        print(row)

# b. Xuất dòng k
def xuat_dong_k(a, k):
    print("Dòng", k, ":", a[k])

# c. Xuất cột k
def xuat_cot_k(a, k):
    print("Cột", k, ":", [a[i][k] for i in range(len(a))])


# d. Tìm dòng có tổng lớn nhất
def dong_tong_lon_nhat(a):
    max_sum = sum(a[0])
    index = 0

    for i in range(1, len(a)):
        if sum(a[i]) > max_sum:
            max_sum = sum(a[i])
            index = i
    return index, max_sum


# e. Tìm cột có tích nhỏ nhất
def cot_tich_nho_nhat(a):
    m, n = len(a), len(a[0])
    min_tich = None
    index = 0

    for j in range(n):
        tich = 1
        for i in range(m):
            tich *= a[i][j]

        if min_tich is None or tich < min_tich:
            min_tich = tich
            index = j

    return index, min_tich

# f. Xuất phần tử dòng chẵn, cột lẻ
def dong_chan_cot_le(a):
    for i in range(0, len(a), 2):      # dòng chẵn
        for j in range(1, len(a[0]), 2):  # cột lẻ
            print(a[i][j], end=" ")
    print()

# g. TBC số chẵn thuộc dòng lẻ
def tbc_chan_dong_le(a):
    tong = dem = 0
    for i in range(1, len(a), 2):
        for x in a[i]:
            if x % 2 == 0:
                tong += x
                dem += 1
    return tong / dem if dem != 0 else 0

# h. TBC phần tử biên
def tbc_bien(a):
    m, n = len(a), len(a[0])
    tong = dem = 0

    for i in range(m):
        for j in range(n):
            if i == 0 or i == m-1 or j == 0 or j == n-1:
                tong += a[i][j]
                dem += 1

    return tong / dem

# i. Trung bình tích phần tử không thuộc biên
def tb_tich_khong_bien(a):
    m, n = len(a), len(a[0])
    tong_tich = 0
    dem = 0

    for i in range(1, m-1):
        for j in range(1, n-1):
            tong_tich += a[i][j]
            dem += 1

    return tong_tich / dem if dem != 0 else 0

# ================= TEST =================
m, n = 4, 4
a = tao_ma_tran(m, n)
print("========Câu14========")

print("Ma trận:")
xuat_ma_tran(a)

xuat_dong_k(a, 1)
xuat_cot_k(a, 2)

print("Dòng tổng lớn nhất:", dong_tong_lon_nhat(a))
print("Cột tích nhỏ nhất:", cot_tich_nho_nhat(a))

print("Dòng chẵn cột lẻ:")
dong_chan_cot_le(a)

print("TBC số chẵn dòng lẻ:", tbc_chan_dong_le(a))
print("TBC phần tử biên:", tbc_bien(a))
print("TBC phần tử không biên:", tb_tich_khong_bien(a))
```

```python
#==================câu15===========================
# a. Xây dựng cấu trúc SINHVIEN (dùng class)
class SinhVien:
    def __init__(self, ma, ten, nam_sinh, dtb):
        self.ma = ma
        self.ten = ten
        self.nam_sinh = nam_sinh
        self.dtb = dtb

    def __str__(self):
        return f"{self.ma} | {self.ten} | {self.nam_sinh} | {self.dtb}"

# b. Tạo danh sách sinh viên
ds = [
    SinhVien("02DH0001", "Phan Lan", 2005, 8.5),
    SinhVien("01CD0002", "Nguyen An", 2004, 4.5),
    SinhVien("03DH0003", "Tran Lan", 2006, 6.0),
    SinhVien("04DH0004", "Phan Binh", 2005, 7.0)
]

# c. Đếm SV đủ điều kiện lên lớp (dtb >= 5)
def dem_len_lop(ds):
    return sum(1 for sv in ds if sv.dtb >= 5)

# d. Xuất SV đủ 20 tuổi
def sv_20_tuoi(ds):
    nam_hien_tai = 2025
    for sv in ds:
        if nam_hien_tai - sv.nam_sinh == 20:
            print(sv)

# e. Đếm SV hệ DH (ký tự 2,3 là "DH")
def dem_he_DH(ds):
    return sum(1 for sv in ds if sv.ma[2:4] == "DH")

# f. Đếm SV tên "Lan"
def dem_ten_Lan(ds):
    return sum(1 for sv in ds if "Lan" in sv.ten)

# g. Đếm SV họ "Phan"
def dem_ho_Phan(ds):
    return sum(1 for sv in ds if sv.ten.startswith("Phan"))

# ================= TEST =================
print("Số SV lên lớp:", dem_len_lop(ds))

print("\nSV đủ 20 tuổi:")
sv_20_tuoi(ds)

print("\nSố SV hệ DH:", dem_he_DH(ds))
print("Số SV tên Lan:", dem_ten_Lan(ds))
print("Số SV họ Phan:", dem_ho_Phan(ds))
```
